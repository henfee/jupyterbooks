</p><div class="section" title="Notebook format"><div class="titlepage"><div><div><h2 class="title"><a id="ch07lvl2sec42"/>Notebook format</h2></div></div></div><p>The notebook format (IPYNB) is the native format for your notebook. We have looked into this IPYNB file in earlier chapters to see what Jupyter is storing for your notebook.</p><p>You would use the notebook format if you wanted to give another user complete access to your notebook since they would run your notebook from their system.</p><p>You may also want to do this to save your notebook in another medium.</p></div><div class="section" title="JavaScript format"><div class="titlepage"><div><div><h2 class="title"><a id="ch07lvl2sec43"/>JavaScript format</h2></div></div></div><p>JavaScript (<code class="literal">.js</code>) format corresponds to the JavaScript implementation of your notebook. If you used JavaScript as the language for your notebook, this is a direct export of the notebook page.</p><p>If you used another language for the script of the notebook, such as Python, then the <span class="strong"><strong>Download&#xA0;as</strong></span> option would change appropriately, that is, <span class="strong"><strong>Download&#xA0;as</strong></span>&#xA0;|&#xA0;<span class="strong"><strong>Python</strong></span>
<span class="strong"><strong>(.py)</strong></span>.</p><p>Using our example, as expected, the JS format is equivalent to the Jupyter display:</p><pre class="programlisting">const stats = require("stats-analysis"); &#xD;

The generated HTML does not have any coding logic-it only has the HTML necessary to display a similar page.</p><p>The HTML format would be useful to convey to another user the results of your notebook. You may want to do this if you wanted to e-mail your notebook to another user (where the raw HTML would be transported and viewable in an e-mail client application).</p><p>HTML is also useful if you have a web service available where you can insert new pages. If the web server does not have support for Jupyter files (refer to the first section of this chapter), HTML may be your only choice. Similarly, you may not want to hand over your source Jupyter&#xA0;Notebook (IPYNB) file even if the web server does support Jupyter.</p><p>The exported HTML format looks like this in a browser:</p><p>

Otherwise, it does look the same as the Jupyter display.</p></div><div class="section" title="Markdown format"><div class="titlepage"><div><div><h2 class="title"><a id="ch07lvl2sec45"/>Markdown format</h2></div></div></div><p>Markdown (<code class="literal">.md</code>) format is a looser version of HTML (remember that HTML stands for Hypertext Markup Language).
