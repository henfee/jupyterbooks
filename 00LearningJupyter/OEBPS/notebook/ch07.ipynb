There is a typical mechanism available for sharing that we will cover in this chapter-placing your notebook on an accessible server on the Internet.</p><p>When you provide a notebook to another person, they may need the notebook in a different format given their system requirements. We will also cover some mechanisms available for providing your notebook to others in a different format as well.</p><p>In this chapter, we will cover the following topics:</p><div class="itemizedlist"><ul class="itemizedlist"><li class="listitem" style="list-style-type: disc">Sharing notebooks</li><li class="listitem" style="list-style-type: disc">Converting notebooks</li></ul></div><div class="section" title="Sharing notebooks"><div class="titlepage"><div><div><h1 class="title"><a id="ch07lvl1sec58"/>Sharing notebooks</h1></div></div></div><p>The typical mechanism for sharing notebooks is to provide your notebook on a website. A website runs on a server or on allocated machine space. The server takes care of all the bookkeeping involved in running a website, such as keeping track of multiple users and logging people on and off.</p><p>In order for the notebook to be of use, though, the website must have notebook logic installed. A typical website knows how to deliver content as HTML given some source files. The most basic form is pure HTML, where every page you access on the website corresponds exactly to one HTML file on the web server. Other languages could be used to develop the website (such as Java or PHP), so then the server needs to know how to access the HTML it needs from those source files. In our context, the server needs to know how to access your notebook in order to deliver HTML to users.</p><p>Even when notebooks are just running on your local machine, they are running in a browser that is accessing your local machine (server) instead of the Internet-so the Web, HTML, and Internet access has already been provided.</p><p>If a notebook is on true website, it is available to everyone who can access that website-whether the server is running on your machine in an office environment accessible over the local area network or if your website is available to all users over the Internet.</p><p>You can always add security around the website so that the only people who can use your notebook are given access by you. Security mechanisms depend on the type of web server software involved.</p><div class="section" title="Sharing notebooks on a notebook server"><div class="titlepage"><div><div><h2 class="title"><a id="ch07lvl2sec38"/>Sharing notebooks on a notebook server</h2></div></div></div><p>Built into the Jupyter process is the ability to expose the notebook as its own web server. Assuming the server is a machine accessible by other users, you can configure Jupyter to run on that server. You must provide the configuration information to Jupyter so it knows how to proceed. The command to generate a configuration file for your Jupyter installation is as follows:</p><pre class="programlisting">
<span class="strong"><strong>jupyter notebook -generate-config</strong></span>
</pre><p>This command will generate a <code class="literal">jupyter_notebook_config.py</code> file in your <code class="literal">~./jupyter</code> directory. For Microsoft users, that directory is a subdirectory of your home user directory.</p><p>The configuration file contains the settings that you can use to expose your notebook as a server:</p><pre class="programlisting">c.NotebookApp.certfile = u'/path/to/your/cert/cert.pem' &#xD;
c.NotebookApp.keyfile = u'/ path/to/your/cert/key.key' &#xD;
c.NotebookApp.ip = '*' &#xD;
c.NotebookApp.password = u'hashed-password' &#xD;
c.NotebookApp.open_browser = False &#xD;
c.NotebookApp.port = 8888 &#xD;
</pre><p>The settings in the file are explained in this table:</p><div class="informaltable"><table border="1"><colgroup><col/><col/></colgroup><tbody><tr><td>

An IP address is a four-part number that identifies the locale of the server involved. An IP address might look like <code class="literal">172.32.88.7</code>.</p><p>Web browsers in concert with Internet software know how to use the IP address to locate the server of interest. The set of software also knows how to translate the URL you mentioned in your browser, such as 
<a class="ulink" href="http://www.microsoft.com">http://www.microsoft.com</a>
, into an IP address.</p></div></div><p>Note that, the example configuration provided is not&#xA0;enterprise ready. You need to coordinate with your security personnel to configure correctly. Once you have changed the settings appropriately, you should be able to point a browser at the URL configured and access your notebook. The URL would be the concatenation of either HTTP or HTTPS (depending on whether you applied an SSL certificate), the IP address, and the port, for example, <code class="literal">HTTPS://123.45.56.9:8888</code>.</p><div class="section" title="Encrypted sharing notebooks on a notebook server"><div class="titlepage"><div><div><h3 class="title"><a id="ch07lvl3sec7"/>Encrypted sharing notebooks on a notebook server</h3></div></div></div><p>Two of the preceding settings can be used if you have an SSL certificate to apply. Without the SSL certificate, the password (refer to the previous section) and all other interactions will be transmitted from the user's browser to the server in the clear. If you are dealing with sensitive information in your notebook, you should obtain an SSL certificate and make the appropriate settings changes for your server.</p><p>If you need more security for access to your notebook, the next step would be to provide an SSL certificate (placed on your machine and the path provided in the configuration file). There are a number of companies that provide SSL certificates. The cheapest at the time of writing is Let's Encrypt, which will provide a low-level SSL certificate for free. (There are gradations of SSL certificates that are not free.)</p><p>Again, once you have set the preceding settings with regard to your certificate, you should be able to access your notebook server using the <code class="literal">https://</code> prefix-knowing that all the transmissions between the user's browser and the notebook server are encrypted and therefore secure.&#xA0;You may need to add your certificate to the keychain on the host machine before browser authentication will work.</p></div></div><div class="section" title="Sharing notebooks on a web server"><div class="titlepage"><div><div><h2 class="title"><a id="ch07lvl2sec39"/>Sharing notebooks on a web server</h2></div></div></div><p>In order to add your notebook to an existing web server, you need to take the preceding steps and add a little more information to the notebook configuration file:</p><pre class="programlisting">c.NotebookApp.tornado_settings = { &#xD;
    'headers': { &#xD;
        'Content-Security-Policy': "frame-ancestors 'https://yourwebsite.com' 'self' " &#xD;
    } &#xD;
} &#xD;
</pre><p>Here, you replace <code class="literal">yourwebsite.com</code> with the URL of your website.</p><p>Once complete, you can access the notebook through your website.</p></div><div class="section" title="Sharing notebooks through Docker"><div class="titlepage"><div><div><h2 class="title"><a id="ch07lvl2sec40"/>Sharing notebooks through Docker</h2></div></div></div><p>Docker is an open lightweight container for distributing software. A typical Docker instance has an entire web server and a specific web application running on a port in a machine. The specifics about the software running in a Docker instance are governed by the <code class="literal">Dockerfile</code> file. This file provides commands to the Docker environment regarding which components to use to configure this instance. Sample <code class="literal">Dockerfile</code> contents for a Jupyter implementation are as follows:</p><pre class="programlisting">ENV TINI_VERSION v0.6.0 &#xD;
ADD https://github.com/krallin/tini/releases/download/${TINI_VERSION}/tini /usr/bin/tini &#xD;
RUN chmod +x /usr/bin/tini &#xD;
ENTRYPOINT ["/usr/bin/tini", "--"] &#xD;
EXPOSE 8888 &#xD;
CMD ["jupyter", "notebook", "-port=8888", "-no-browser", "-ip=0.0.0.0"] &#xD;
</pre><p>Here is a discussion about each of the commands in the <code class="literal">Dockerfile</code>:</p><div class="itemizedlist"><ul class="itemizedlist"><li class="listitem" style="list-style-type: disc">The <code class="literal">ENV</code> command tells Docker to use a specialized operating system. This is necessary to overcome a deficiency of Jupyter that keeps obtaining and releasing resources from your machine.&#xA0;TINI is third-party software that provides a minimum Docker initialization.</li><li class="listitem" style="list-style-type: disc">The <code class="literal">ADD</code> command tells Docker where the <code class="literal">tini</code> code is located.</li><li class="listitem" style="list-style-type: disc">The <code class="literal">RUN</code> command is changing the access rights to the <code class="literal">tini</code> directory.</li><li class="listitem" style="list-style-type: disc">The <code class="literal">ENTRYPOINT</code> command tells Docker what to use as the operating system of the Docker instance.</li><li class="listitem" style="list-style-type: disc">The <code class="literal">EXPOSE</code> command tells Docker which port to expose your notebook on.</li><li class="listitem" style="list-style-type: disc">The <code class="literal">CMD</code> command tells Docker which commands to run (once the environment is set up).</li><li class="listitem" style="list-style-type: disc">Once the Docker instance is deployed to your Docker machine, you can access the Docker instance on the machine at the port specified (8888), for example, <code class="literal">http://machinename.com:8888</code>.</li></ul></div><p>The instructions mentioned previously assume you are new to Docker. If you have an existing Docker installation the Dockerfile configuration and access to your Jupyter instance in Docker, the instructions would probably change.</p></div><div class="section" title="Sharing notebooks on a public server"><div class="titlepage"><div><div><h2 class="title"><a id="ch07lvl2sec41"/>Sharing notebooks on a public server</h2></div></div></div><p>Currently, one of the&#xA0;hosting companies that allows you to host your notebook(s) for free is GitHub. For example, in prior chapters, to use R programming in your notebook you would have had to install the R tool set on your machine.
