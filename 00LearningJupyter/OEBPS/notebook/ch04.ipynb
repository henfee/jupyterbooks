There are separate steps for making Julia scripting available in your Jupyter installation.</p><div class="section" title="Adding Julia scripts to Jupyter on a Mac"><div class="titlepage"><div><div><h2 class="title"><a id="ch04lvl2sec16"/>Adding Julia scripts to Jupyter on a Mac</h2></div></div></div><p>If you are running on a Mac, you are in luck. The Mac installation of Jupyter includes Julia 0.4.5, as can be seen in the <span class="strong"><strong>New</strong></span> menu:</p><p>
</p><div class="mediaobject"><img src="graphics/image_04_001.jpg" alt="Adding Julia scripts to Jupyter on a Mac"/></div><p>
</p></div><div class="section" title="Adding Julia scripts to Jupyter on Windows"><div class="titlepage"><div><div><h2 class="title"><a id="ch04lvl2sec17"/>Adding Julia scripts to Jupyter on Windows</h2></div></div></div><p>If you are running on a Windows machine, there are a few steps to get Julia on Jupyter. Remember, this environment was really developed for Linux.</p><p>First, we need to install Julia on your Windows machine. Navigate to the Julia download page (<a class="ulink" href="http://julialang.org/downloads/">http://julialang.org/downloads/</a>), download the correct version, which is Julia 0.4.5 for Windows (in my case, I used the Windows self-extracting EXE file for 32-bit machines), and run the installation with standard defaults.</p><div class="note" title="Note" style=""><div class="inner"><h3 class="title"><a id="note13"/>Note</h3><p>You must run the Julia installation as Administrator on your machine. After downloading the file, open the <code class="literal">Downloads</code> folder, right-click on the Julia executable, and select <span class="strong"><strong>Run as administrator</strong></span>.</p></div></div><p>Once the install is complete, you should verify that everything worked. Select Julia from the programs list and run the Julia program. You should see the Julia command-line, as shown in the following screenshot:</p><p>
</p><div class="mediaobject"><img src="graphics/image_04_002.jpg" alt="Adding Julia scripts to Jupyter on Windows"/></div><p>
</p><p>In the Julia window, run the following command to add Julia to Jupyter:</p><pre class="programlisting">
<span class="strong"><strong>Pkg.add("IJulia")</strong></span>
</pre><p>This will result in <span class="strong"><strong>IJulia</strong></span> (Interactive Julia) being downloaded and installed on your machine. There are many sub-packages that will get automatically installed as well. Your display will look something like the following screenshot:</p><p>
</p><div class="mediaobject"><img src="graphics/image_04_003.jpg" alt="Adding Julia scripts to Jupyter on Windows"/></div><p>
</p><p>Julia uses font color as feedback. I entered the text <code class="literal">Pkg.add</code> in white at the top of the screen, successful execution steps are in blue, and possible problems are shown in red. You must wait for the installation to complete. The last line should have read as follows:</p><pre class="programlisting">
<span class="strong"><strong>INFO: Package database updated</strong></span>
</pre><p>At this point, you can close the Julia window (using the <code class="literal">quit()</code> command).</p><p>One last step is to open your notebook (using the <code class="literal">jupyter notebook</code> command), and if you open the <span class="strong"><strong>New</strong></span> menu (in the upper-right corner of the screen), you should see a Julia type available as shown in the following screenshot:</p><p>
</p><div class="mediaobject"><img src="graphics/image_04_004.jpg" alt="Adding Julia scripts to Jupyter on Windows"/></div><p>
</p></div><div class="section" title="Adding Julia packages to Jupyter"><div class="titlepage"><div><div><h2 class="title"><a id="ch04lvl2sec18"/>Adding Julia packages to Jupyter</h2></div></div></div><p>The standard installation for Julia&#xA0;in Jupyter has many packages that are commonly used in Julia programming. However, if you do need to add another package, it is a small number of steps:</p><div class="orderedlist"><ol class="orderedlist arabic"><li class="listitem">Close down your notebook (including the server).</li><li class="listitem">Run the Julia command-line program and type the following:<pre class="programlisting">
<span class="strong"><strong>Pkg.add("DataFrames")</strong></span>
<span class="strong"><strong>Pkg.add("RDatasets")</strong></span>
<span class="strong"><strong>Pkg.add("Gadfly")</strong></span>
<span class="strong"><strong>quit();</strong></span>
</pre></li><li class="listitem">Restart your notebook and the package should be available in your Julia script. To use an installed package, type <code class="literal">library</code>(name of the R package you want to add).</li></ol></div><p>I would recommend adding the three preceding packages right away as they are needed for many scripts.</p><div class="note" title="Note" style=""><div class="inner"><h3 class="title"><a id="note14"/>Note</h3><p>The first time you use a package in Julia you will see a line highlighted in light red that shows Julia is precompiling, such as this:</p></div></div><pre class="programlisting">
<span class="strong"><strong>INFO: Precompiling module Dataframes...</strong></span>
</pre><p>You can use the <code class="literal">Pkg.add(&#x2026;)</code> function directly in your script, but that doesn't seem correct. Every time you run your script, the system will attempt to validate whether you have the specified package or not, install it into your environment if needed, and even tell you if you are out of date. None of these steps belong as part of your script.</p></div></div></div></body></html>
