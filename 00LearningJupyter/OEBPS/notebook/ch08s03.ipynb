<html xmlns="http://www.w3.org/1999/xhtml"><head><title>Docker</title><link rel="stylesheet" href="epub.css" type="text/css"/><meta name="generator" content="DocBook XSL Stylesheets V1.75.2"/></head><body id="page"><div class="section" title="Docker"><div class="titlepage"><div><div><h1 class="title"><a id="ch08lvl1sec63"/>Docker</h1></div></div></div><p>Docker is another mechanism that can be used to allow multiple users of the same notebook without collision. Docker is a system that allows you to construct sets of applications into an <span class="emphasis"><em>image</em></span> that can be run in a container. It runs in most environments. Docker allows for many instances of an image to be run in the same machine, but maintains separate address space, so each user of a Docker image has their own instance of the software and their own set of data/variables.</p><p>Each image is a complete stack of software necessary to run, for example, a web server, web application(s), API(s), and so on.</p><p>It is not a large leap to think of an image of your notebook. The image contains Jupyter server code and your notebook. The result is a completely intact unit that does not share any space with another's instance.</p><div class="section" title="Installation"><div class="titlepage"><div><div><h2 class="title"><a id="ch08lvl2sec52"/>Installation</h2></div></div></div><p>Installing Docker involves downloading the latest version (the <code class="literal">docker.dmg</code> file for a Mac and the EXE install for Windows) and then copying the Docker applications into your Applications folder. Older Mac editions may need to use Docker toolbox instead.</p><p>
<span class="strong"><strong>Docker QuickStart Terminal</strong></span> is the go-to application for most developers. Docker QuickStart will start Docker on your local machine, allocate an IP address/port for addressing the Docker application(s), and bring you into the Docker terminal. Once QuickStart has completed, if you have installed your image, you could access your application (in this case your Jupyter&#xA0;Notebook).</p><p>From the Docker terminal, you can load images, remove images, check status, and so on.</p></div><div class="section" title="Starting Docker"><div class="titlepage"><div><div><h2 class="title"><a id="ch08lvl2sec53"/>Starting Docker</h2></div></div></div><p>If you run Docker Quickstart, you will be brought to the Docker terminal window with a display like this:</p><pre class="programlisting">

We need to build an image with a notebook and place it in Docker.</p><p>We need to download all of the Jupyter-Docker coding necessary. In the Docker terminal window, we run the <code class="literal">docker pull jupyter/all-spark-notebook</code> command:</p><pre class="programlisting">
<span class="strong"><strong>docker pull jupyter/all-spark-notebook</strong></span>
<span class="strong"><strong>Using default tag: latest</strong></span>
<span class="strong"><strong>latest: Pulling from jupyter/all-spark-notebook</strong></span>
<span class="strong"><strong>8b87079b7a06: Pulling fs layer </strong></span>
<span class="strong"><strong>872e508604af: Pulling fs layer </strong></span>
<span class="strong"><strong>8e8d83eda71c: Pull complete </strong></span>
<span class="strong"><strong>...</strong></span>
</pre><p>This is a large download that will take some time. It is downloading and installing all of the possibly necessary components needed to run Jupyter in an image. Remember, each image is completely self-contained, so each image has <span class="emphasis"><em>all</em></span> of the software needed to run Jupyter.</p><p>Once the download is complete, we can start an image for our notebook using a command such as the following:</p><pre class="programlisting">
<span class="strong"><strong>docker run -d -p 8888:8888 -v /disk-directory:/virtual-notebook jupyter/all-spark-notebook</strong></span>
</pre><p>The parts of this command are as follows:</p><div class="itemizedlist"><ul class="itemizedlist"><li class="listitem" style="list-style-type: disc"><code class="literal">docker run</code>: The command to Docker to start executing an image.</li><li class="listitem" style="list-style-type: disc"><code class="literal">-d</code>: Run the image as a server (daemon) that will continue running until manually stopped by the user.</li><li class="listitem" style="list-style-type: disc"><code class="literal">-p 8888:8888</code>: Expose the internal port <code class="literal">8888</code> to external users with the same port address. Notebooks use port <code class="literal">8888</code> by default already, so we are saying just expose the same port.</li><li class="listitem" style="list-style-type: disc"><code class="literal">-v /disk-directory:/virtual-notebook</code>: Take the notebook from the disk directory and expose it as the virtual notebook name.</li><li class="listitem" style="list-style-type: disc">The last argument is to use the <code class="literal">all-spark-notebook</code> as the basis for this image.</li></ul></div><p>In my case, I used the following command:</p><pre class="programlisting">
<span class="strong"><strong>$ docker run -d -p 8888:8888 -v /Users/dtoomey:&#xD;
    /dan-notebook jupyter/all-spark-notebook</strong></span>
<span class="strong"><strong>b59eaf0cae67506e4f475a9861f61c01c5af3556489992104c4ce39343e8eb02</strong></span>
</pre><p>The big hex number displayed is the image identifier.</p><p>We can make sure the image is running using the <code class="literal">docker ps -l</code> command, which lists out the images in our <code class="literal">docker</code>:</p><pre class="programlisting">
<span class="strong"><strong>$ docker ps -l</strong></span>
<span class="strong"><strong>CONTAINER ID        IMAGE                        COMMAND                 &#xD;
    CREATED             STATUS               PORTS                       &#xD;
    NAMES</strong></span>
<span class="strong"><strong>b59eaf0cae67        jupyter/all-spark-notebook   "tini -- start-&#xD;
    notebo"   8 seconds ago       Up 7 seconds        0.0.0.0:8888-&#xD;
    &gt;8888/tcp   modest_bardeen</strong></span>
</pre><p>The parts of the display are as follows:</p><div class="itemizedlist"><ul class="itemizedlist"><li class="listitem" style="list-style-type: disc">The first name,&#xA0;<code class="literal">b59eaf0cae67</code>, is the assigned <code class="literal">ID</code> of the container. Each image in Docker is assigned to a container.</li><li class="listitem" style="list-style-type: disc">The <code class="literal">IMAGE</code> is <code class="literal">Jupyter/all-spark-notebook</code>-it contains all of the software needed to run your notebook.</li><li class="listitem" style="list-style-type: disc">The command is telling Docker to start the image.</li><li class="listitem" style="list-style-type: disc">The port access point is as we expected: <code class="literal">8888</code>.</li><li class="listitem" style="list-style-type: disc">Lastly, Docker assigns random names to every running image; ours is&#xA0;<code class="literal">modest bardeen</code> (I'm not sure why they do this).</li></ul></div><p>At this point, we should be able to access the notebook from an external browser at <code class="literal">http:// 192.168.99.100:8888</code>.
