However, if you do need to add another package, it is a small number of steps:</p><div class="orderedlist"><ol class="orderedlist arabic"><li class="listitem">Close down your notebook (including the server).</li><li class="listitem">In the command-line window, type the following:<pre class="programlisting">
<span class="strong"><strong>R</strong></span>
<span class="strong"><strong>install.packages("name of the R package you want to add")</strong></span>
<span class="strong"><strong>quit()</strong></span>
<span class="strong"><strong># answer Yes to save</strong></span>
</pre></li><li class="listitem">Restart your notebook, and the package should be available for use in your R script. To use an installed package, type <code class="literal">library</code>(name of the R package you want to add).</li></ol></div><p>Note, you may still have problems in R if the core version of R that you have installed is out of date and you need to upgrade to use a particular library.</p></div><div class="section" title="R limitations in Jupyter"><div class="titlepage"><div><div><h2 class="title"><a id="ch03lvl2sec12"/>R limitations in Jupyter</h2></div></div></div><p>In this chapter, we used a variety of packages, both pre-installed and installed especially for the example. I have exercised a variety of materials available in R under Jupyter and have not found any limitations; you can do most of the steps in Jupyter that you would have done under the standard R implementations. The only limitation appears to be when you are using <span class="strong"><strong>Shiny</strong></span> or if you are attempting to use extensive markdown:</p><div class="itemizedlist"><ul class="itemizedlist"><li class="listitem" style="list-style-type: disc">For Shiny, I think you are mixing purposes-Jupyter&#xA0;provides a web experience and so does Shiny-so I'm not sure how to even decide if this should work. This issue is being addressed by the Jupyter development group.</li><li class="listitem" style="list-style-type: disc">Using extensive markdown also does not appear to be a good idea. The intent of markdown was to allow notebook developers to augment the standard output (of R) in a more illustrative manner. I think if you are adding extensive markdown to your notebook you really need to develop a website, maybe using Shiny-then you would have all HTML markdown available.</li></ul></div></div><div class="section" title="After adding R scripts to Jupyter"><div class="titlepage"><div><div><h2 class="title"><a id="ch03lvl2sec13"/>After adding R scripts to Jupyter</h2></div></div></div><p>Once the notebook is up and running (you should be able to use the command-line <code class="literal">jupyter notebook</code>), if you open the <span class="strong"><strong>New</strong></span> menu in the top-right, you will see a choice has been added for R scripting:</p><p>
